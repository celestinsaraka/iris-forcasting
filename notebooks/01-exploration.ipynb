# Iris Dataset Exploration

In this notebook, we will perform exploratory data analysis (EDA) on the Iris dataset. We will visualize the data and gain initial insights.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris_data = pd.read_csv('../data/raw/iris.csv')
iris_data.head()
```

```python
# Summary statistics
iris_data.describe()
```

```python
# Visualizing the distribution of species
sns.countplot(x='species', data=iris_data)
plt.title('Distribution of Iris Species')
plt.show()
```

```python
# Pairplot to visualize relationships
sns.pairplot(iris_data, hue='species')
plt.title('Pairplot of Iris Features')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(iris_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Iris Features')
plt.show()
```
